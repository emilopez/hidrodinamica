## La dinámica de la vorticidad está dada por

$$\partial _t \vec{\omega} +  (\vec{u} \cdot \nabla)\vec{\omega} = (\vec{\omega} \cdot \nabla)\vec{u} + \nu \nabla ^2\vec{\omega}$$

Para el problema visto en clase donde suponemos una vorticidad 1D tal que $\vec{\omega} = (\omega, 0, 0)$ y $\vec{\omega} = \omega(r,t)$

$$\vec{u} = (u_x, u_r, u_\phi)$$
$$\vec{\omega} \cdot \nabla\vec{u} = 0$$

### Continuidad

La ecuación de continuidad en coordenadas cilíndricas está dada por:

$$\nabla \cdot u = \frac{1}{r}\partial_r(ru_r) + \frac{1}{r}\partial_\phi(u_\phi) + \partial_z(u_z)$$

Para el problema dado donde $u_x = \alpha x$; $u_r = -\alpha \frac{r}{2}$; $u_\phi = ?$

Calculamos $$\nabla \cdot u = \partial_x[\alpha x] + \frac{1}{r} \partial_r[r(-\alpha \frac{r}{2})] + \frac{1}{r}\partial_\phi [u_\phi]$$

y observamos que satisface continuidad $\nabla \cdot u = 0$ ya que:

$$ \partial_x[\alpha x] = \alpha $$
$$ \partial_r\left[r\left(-\alpha \frac{r}{2}\right)\right] = -\alpha r$$
$$\partial_\phi [u_\phi] = 0$$

Aglutinando la ecuación nos queda:

$$ \alpha + \frac{1}{r} (-\alpha) r = \alpha - \alpha = 0 $$

Por lo que se satisface continuidad. Teniendo en cuenta que se considera un flujo incompresible, la conservación de la cantidad de movimiento también es nula.

### Resolución de los términos de la Ec. de vorticidad

El operador $\nabla$ en coordenadas cilíndricas está dado por:

$$\nabla = \frac{\partial}{\partial_r}\vec{r} + 
\frac{1}{r}\frac{\partial}{\partial_ \phi} \vec{\phi}+
\frac{\partial}{\partial_z}\vec{z}
$$

- Resolvemos $(\vec{u} \cdot \nabla )w$:

$$(\vec{u} \cdot \nabla )w = \bigg(u_x, u_r, u_\alpha \bigg) \cdot \left(\partial_x, \partial_r, \frac{1}{r}\partial_\phi\right) \omega \vec{e_x}$$

$$=\bigg(u_x \partial _x + u_r \partial_r + \frac{1}{r} \partial_\phi \bigg)\omega \vec{e_x}$$

ya que $\omega = \omega(r,t)$:

$$ = u_r \frac{dw}{dr} \vec{e_x}$$

- Resolvemos $\vec{\omega} \cdot \nabla\vec{u}$

$$\vec{\omega} \cdot \nabla\vec{u} = \bigg(\omega, 0, 0\bigg) \cdot \bigg(\partial_x, \partial_r, \partial_\alpha \bigg)\vec{u}$$

$$ = \omega \partial_x \vec{u} $$
$$ = \omega \partial_x \bigg[\alpha x \vec{e_x} - \frac{\alpha r}{2} \vec{e_r} \bigg]$$
$$ = \alpha \omega \vec{e_x}$$

- Resolvemos $\nu \nabla ^2\vec{\omega}$

$$\nu \nabla ^2\vec{\omega} = \nu \bigg[ \frac{\partial^2\omega}{\partial x^2} + \frac{1}{r} \frac{r\partial \omega}{\partial r} + \frac{1}{r^2} \frac{\partial^2 \omega}{\partial \phi^2} \bigg]\vec{e_x}$$

El primer y último término se anulan ya que $\vec{\omega} = \omega(r,t)$, por lo tanto:

$$=  \frac{\nu}{r}\frac{d}{dr}\bigg(r\frac{d\omega}{dr}\bigg)\vec{e_x}$$


Combinando los últimos items resueltos en la ecuación de vorticidad, nos queda:

$$\partial _t \vec{\omega} +  u_r \frac{d\omega}{dr} \vec{e_x} = \alpha \omega \vec{e_x} +  \frac{\nu}{r}\frac{d}{dr}\bigg(r\frac{d\omega}{dr}\bigg)\vec{e_x}$$

Eliminando el término $\partial _t \vec{\omega}$ debido a que es un flujo estacionario y reemplazando $u_r = -\frac{\alpha r}{2}$ nos queda para la componente $\vec{e_x}$:

$$\frac{\alpha r}{2} \frac{d\omega}{dr} + \alpha \omega +  \frac{\nu}{r}\frac{d}{dr}\bigg(r\frac{d\omega}{dr}\bigg) = 0$$

$$\alpha \bigg( \omega + \frac{r}{2} \frac{d\omega}{dr} \bigg)+  \frac{\nu}{r}\frac{d}{dr}\bigg(r\frac{d\omega}{dr}\bigg) = 0$$

Teniendo en cuenta que el primer término es la regla del producto de derivar $\frac{1}{2r}\frac{d}{dr}(r^2\omega)$, reacomodamos sacando como factor común $\frac{1}{r}$

$$ \frac{1}{r} \bigg[ \frac{\alpha}{2} \frac{d}{dr}\bigg(r^2\omega \bigg) +  \frac{\nu}{r}\frac{d}{dr}\bigg(r\frac{d\omega}{dr}\bigg) \bigg] = 0$$

Integrando lo que se encuentra entre corchetes obtenemos:

$$\frac{\alpha}{2} r^2\omega +  \nu r \frac{d\omega}{dr} = C_1$$

Haciendo $C_1 = 0$, reacomodando, nos queda

$$ \nu r \frac{d\omega}{dr} = -\frac{\alpha}{2} r^2\omega$$

Dividiendo ambos miembros por $r\omega$ y separando variables nos queda integrar

$$ \int \frac{1}{\omega}d\omega = -\frac{\alpha}{2\nu} \int r dr$$

$$ ln \bigg(\frac{\omega}{\omega_0} \bigg)= \frac{-\alpha}{2\nu} \frac{r^2}{2}$$

$$\omega = \omega_0 e^{-\alpha r^2 /4 \nu}$$

Visualizamos la distribución de vorticidad en el siguiente gráfico

```python
plt.figure()
r = np.arange(-2.5,2.5,0.1)
w0 = 2
w = w0 * np.exp(-(alpha*r**2)/(4*nu))
plt.plot(r,w)
plt.show()
```

Para obtener la componenente de la velocidad $u_\phi$ plateamos:

$$\vec{\omega} = \nabla \times \vec{u}$$
$$ = \vec{e_x} \bigg( \frac{1}{r} \frac{\partial r}{\partial r} u_\phi - \frac{1}{r} \frac{\partial u_r}{\partial \phi} \bigg) +
\vec{e_r} \bigg( \frac{1}{r} \frac{\partial u_x}{\partial \phi} - \frac{\partial u_\phi}{\partial x} \bigg) +
\vec{e_\phi} \bigg( \frac{u_r}{u_x} - \frac{1}{r}\frac{\partial u_x}{\partial r} \bigg)$$

Debido a que la única componente no nula de la vorticidad es $\vec{\omega} = \omega \vec{e_x}$ y que $u_x = \alpha x$ y $u_r = -\frac{\alpha r}{2}$, nos queda:

$$\vec{\omega} = \vec{e_x} \frac{1}{r}\frac{\partial r u_\phi}{\partial r} $$

Igualando al resultado obtenido previamente (distribución gaussiana de vorticidad) tenemos:

$$\omega = \frac{1}{r}\frac{d (r u_\phi)}{d r}  = \omega_0 e^{-\alpha r^2 /4 \nu}$$

Reorganizando nos queda la ecuación diferencial:

$$d (r u_\phi)  = \omega_0 r e^{-\alpha r^2 /4 \nu} dr$$

Aplicando integración a ambos lados obtenemos:

$$u_\phi  = \frac{\omega_0}{r} \int_0 ^r r e^{-\alpha r^2 /4 \nu} dr$$

Aplicamos un cambio de variable para resolver la integral previa, haciendo:

$$u = -\frac{\alpha}{4 \nu} r^2$$
$$du = -\frac{\alpha}{2 \nu} r$$

Reemplazando, la integral queda del siguiente modo:
$$ \frac{-2\nu}{\alpha}\int e^u du =$$

Volviendo al valor de $u = -\frac{\alpha}{4 \nu} r^2$

$$ \frac{-2\nu}{\alpha} \bigg[e^{-\frac{\alpha}{4 \nu} r^2} \bigg]_0^r =$$

$$ =\frac{-2\nu}{\alpha} \bigg[e^{-\frac{\alpha}{4 \nu} r^2} -1 \bigg]$$


Incorporamos estos resultados en la ecuación de $u_\phi$ obtenemos:

$$u_\phi  = \frac{2 \nu \omega_0}{\alpha r} \bigg[1- e^{-\alpha r^2 /4 \nu} \bigg]$$


## Visualización

Teniendo las tres componenetes de velocidad $u_x$, $u_r$ y $u_\phi$ visualizamos el campo vectorial haciendo las transformaciones a coordenadas cartesianas correspondientes. A continuación el código y su visualización correspondiente.

```python
%matplotlib notebook

from mpl_toolkits.mplot3d import axes3d
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')

# Grilla espacial
x, y, z = np.meshgrid(np.arange(0.1, 2, 0.2), np.arange(0.1, 2, 0.2), np.arange(0.1, 2, 0.5))
r = np.sqrt(x**2 + y**2)

# Parametros
alpha = 1
nu    = 0.25
w0    = 2

# Make the direction data for the arrows
ux    = alpha * x
ur    = -(alpha/2) * r
uphi  = ((2 * nu * w0) / (alpha * r)) * (1-np.exp(-(alpha*r**2)/(4*nu)))

u = ux
v = r*np.sin(uphi)
w = r*np.cos(uphi)

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
